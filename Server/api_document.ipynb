# 사용 가이드: 모델 예측 결과 출력 API (Postman 활용)   
모델은 따로 실행 후 저장해서 사용할 것

## **1. 코드 실행**
- **Colab 환경**에서 Flask 서버를 실행하는 경우, `flask-ngrok`을 통해 외부에서 접근할 수 있도록 설정합니다.
  - `flask-ngrok` 설치 후 실행하면, 생성된 ngrok URL이 제공됩니다:
    ```
    Running on http://<ngrok-url>
    ```
  - Postman에서 이 ngrok URL을 사용합니다.
- **로컬 환경**에서 실행하는 경우:
  - 기본적으로 Flask 서버는 `http://127.0.0.1:5000`에서 실행됩니다.
  - 로컬 네트워크에서 접근하려면 `http://<local-ip>:5000`을 사용할 수 있습니다.

## **2. Postman으로 API 테스트**

- **새 요청 생성**:
  1. Postman에서 **New** > **HTTP Request** 선택.
  2. Request URL 입력: 
     - Colab 환경: `http://<ngrok-url>/predict`
     - 로컬 환경: `http://127.0.0.1:5000/predict`
  3. **HTTP Method**: `POST`로 설정.
  4. **Headers** 추가:
     - `Content-Type`: `application/json`
  5. **Body 설정**: `raw` 선택 후 JSON 형식으로 입력 데이터 작성:
     ```json
     [
         {"time_idx": 101, "feed_pressure": 50.1, "group": "series"},
         {"time_idx": 102, "feed_pressure": 52.3, "group": "series"}
     ]
     ```

- **요청 전송**:
  - **Send** 버튼을 클릭하여 요청을 전송합니다.
  - 성공적으로 요청이 처리되면 예측 결과가 JSON 형식으로 반환됩니다.
